```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('C://Users//Matth//OneDrive//Documents//Data Science//DATA 557//557_Final_Project//Data//DataScienceData.csv')
pop =  pd.read_csv('C://Users//Matth//OneDrive//Documents//Data Science//DATA 557//557_Final_Project//Data//pop_dense.csv', header=None, encoding="ISO-8859-1")
pop_dense = []
val = pop.values
country_dense = {}
for v in val:
    country_dense[v[0]] = v[1]
for i in range(0,len(data)):
    country = data.iloc[i][3]
    country = str(country)
    if country != "Other" and country != 'nan':
        dense = country_dense[country]
        pop_dense.append(dense)
    else:
        #dense = country_dense[country]
        pop_dense.append(0)
```

```python

density = pd.DataFrame({"Density": pop_dense})
```

```python
df = pd.concat([data, density], axis=1, sort='False')
density.head()
```

```python
df = df.drop(columns="Unnamed: 0")
```

```python
df.to_csv("C://Users//Matth//OneDrive//Documents//Data Science//DATA 557//557_Final_Project//Data//Data_Science_Density.csv")
```
